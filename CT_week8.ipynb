```python
import textwrap

class Calculator:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self):
        return self.x + self.y

    def subtract(self):
        return self.x - self.y

    def product(self):
        return self.x * self.y

    def divide(self):
        if self.y == 0:
            return "Khong the chia cho 0"
        return self.x / self.y

    def power(self):
        return self.x ** self.y

    def mod(self):
        if self.y == 0:
            return "Khong the mod cho 0"
        return self.x % self.y

    def set_numbers(self, a, b):
        self.x = a
        self.y = b

    def show_menu(self):
        print(textwrap.dedent("""\
        Cong : 1
        Tru  : 2
        Nhan : 3
        Chia : 4
        Ham mu : 5
        Mod : 6
        Set : 7 (x va y)
        Thoat : yes
        """))

    def run(self):
        print("calc started")

        while True:
            self.show_menu()
            option = input("ban muon lam gi ? ").lower()

            if option == 'yes':
                print("system out")
                break

            if option not in [str(i) for i in range(1, 8)]:
                print("INVALID OPTION")
                continue

            option = int(option)

            if option == 1:
                print(f"{self.x} + {self.y} = {self.add()}")
            elif option == 2:
                print(f"{self.x} - {self.y} = {self.subtract()}")
            elif option == 3:
                print(f"{self.x} * {self.y} = {self.product()}")
            elif option == 4:
                print(f"{self.x} / {self.y} = {self.divide()}")
            elif option == 5:
                print(f"{self.x} ** {self.y} = {self.power()}")
            elif option == 6:
                print(f"{self.x} mod {self.y} = {self.mod()}")
            elif option == 7:
                print("nhap 2 so x va y cach nhau boi dau cach")
                a, b = map(int, input().split())
                self.set_numbers(a, b)
                print(f"da cap nhat: x = {self.x}, y = {self.y}")
x,y=map(int,input().split())
a=Calculator(x,y)
a.run()
```

```python
#W8A5
import textwrap
class ShoppingCart:
    def __init__(self):
        self.arr =  []
    def add(self,name,price):
        self.arr.append({name:price})
        print("da them san pham")
    def remove_name(self,name):
        for k in self.arr:
            if list(k.keys())[0]== name:
                self.arr.remove(k)
        print(f"da xoa {name} ra khoi list")
    def check_none(self):
        return len(self.arr)==0
    def tong(self):
        tong=0
        for k in self.arr:
            for i in k:
                tong+=int(k[i])
        return tong
    def hien_thisp(self):
        if self.check_none():
            print("ko co sp nao")
        else:
            for k in self.arr:
                print(k)
    def clear(self):
        self.arr=[]
    def show_menu(self):
        print(textwrap.dedent("""\
        -Thêm vào giỏ hàng một sản phẩm mới : 1
        -Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên để xóa)  : 2
        -Kiểm tra xem giỏ hàng có rỗng không : 3
        -Tính tổng tiền giỏ hàng : 4
        -Hiển thị toàn bộ sản phẩm trong giỏ : 5
        -Xóa toàn bộ giỏ hàng : 6
        Thoat : yes
        """))
    def run(self):
        print("calc started")

        while True:
            self.show_menu()
            option = input("ban muon lam gi ? ").lower()

            if option == 'yes':
                print("system out")
                break

            if option not in [str(i) for i in range(1,7)]:
                print("INVALID OPTION")
                continue

            option = int(option)
            if option == 1:
                name=input("nhap ten sp: ").strip()
                while True:
                   price=input("nhap gia san pham: ").strip()
                   if price.isdigit():
                       break
                   else:
                       print("nhap gia hop le")
                self.add(name,price)
            elif option == 2:
                name=input("ban muon xoa sp gi? ").strip()
                self.remove_name(name)
            elif option == 3:
                print("*Hien thi danh sach rong hay ko:")
                if self.check_none():
                    print("danh sach rong")
                else:
                    print("danh sach ko rong ")
            elif option == 4:
                print("*Tong gia la :",self.tong())
            elif option == 5:
                print("*Hien thi danh sach san pham :")
                self.hien_thisp()
            elif option == 6:
                self.clear()
                print("list da duoc lam trong")
a=ShoppingCart()
a.run()
```

```python
#W8A6
class Rectangle:
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return f"{round(self.w*self.h,2):.2f}"
    def perimeter(self):
        return f"{round(self.w*2+2*self.h,2):.2f}"
    def scale(self,k):
        self.w*=k
        self.h*=k
w,h,k=map(int,input().split())
hing=Rectangle(w,h)
hing.scale(k)
print(hing.area())
print(hing.perimeter())
```

```python
#W8A7
from math import gcd
class Fraction:
    def __init__(self,num,den):
        num=int(num)
        den=int(den)
        g=gcd(abs(num),abs(den))
        self.num=num//g
        self.den=den//g
    def __str__(self):
        return f'{self.num}/{self.den}'
    def add(self,other):
        num=self.num*other.den+other.num*self.den
        den=self.den*other.den
        return num,den
    def sub(self,other):
        num=self.num*other.den-other.num*self.den
        den=self.den*other.den
        return num,den
    def mul(self,other):
        num=self.num*other.num
        den=self.den*other.den
        return num,den
    def div(self,other):
        num=self.num*other.den
        den=self.den*other.num
        return num,den
a,b,op,x,y=map(str,input().split())
k=Fraction(x,y)
z=Fraction(a,b)
if op=='+':
    num,den=z.add(k)
elif op=='-':
    num,den=z.sub(k)
elif op=='*':
    num,den=z.mul(k)
else:
    num,den=z.div(k)
g=gcd(abs(num),abs(den))
num=num//g
den=den//g
print(f'{num}/{den}')
```

```python
#W8A8
class BankAccount:
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance
    def __str__(self):
        return f"{self.balance:.2f}"
    def withdraw(self,x):
        if self.balance > x:
            self.balance -= x
    def deposit(self,x):
        self.balance += x
owner, initial_balance=map(str,input().split())
initial_balance=float(initial_balance)
a=BankAccount(owner,initial_balance)  
n=int(input())
for _ in range(n):
    action=input().split()
    if action[0]=='DEPOSIT':
        a.deposit(float(action[1]))
    else:
        a.withdraw(float(action[1]))
print(a)
```

```python
#W8A10
class Book:
    def __init__(self,title, author, year):
        self.title=title
        self.author=author
        self.year=year
class Library:
    def __init__(self):
        self.books=[]
    def add(self,book):
        self.books.append(book)
    def count_by_author(self,name):
        count=0
        for book in self.books:
            if book.author==name:
                count+=1
        return count
    def find_by_year(self,y):
        count=0
        for book in self.books:
            if book.year==y:
                count+=1
        return count
lib=Library()
n=int(input())
for _ in range(n):
    action=(input().strip()).split(' ',1)
    if action[0]=='ADD':
        title,author,year=map(str,action[1].split(';'))
        lib.add(Book(title,author,year))
    elif action[0]=='COUNT':
        print(lib.count_by_author(action[1]))
    elif action[0]=='COUNTYEAR':
        print(lib.find_by_year(action[1]))
```

```python
#W8A11
import math
class Point:
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    def vecto_pt(self, ot1, ot2):
        abx = ot1.x - self.x
        aby = ot1.y - self.y
        abz = ot1.z - self.z

        acx = ot2.x - self.x
        acy = ot2.y - self.y
        acz = ot2.z - self.z

        nx = aby * acz - abz * acy
        ny = abz * acx - abx * acz
        nz = abx * acy - aby * acx

        return (nx, ny, nz)
arr=[]
for k in range(4):
    diem=list(map(float,input().split()))
    arr.append(diem)
A=Point(arr[0][0],arr[0][1],arr[0][2])
B=Point(arr[1][0],arr[1][1],arr[1][2])
C=Point(arr[2][0],arr[2][1],arr[2][2])
D=Point(arr[3][0],arr[3][1],arr[3][2])
m1=A.vecto_pt(B,C)
m2=B.vecto_pt(C,D)
nx1=m1[0]
nx2=m2[0]
ny1=m1[1]
ny2=m2[1]
nz1=m1[2]
nz2=m2[2]
dot = nx1*nx2 + ny1*ny2 + nz1*nz2
len1 = math.sqrt(nx1**2 + ny1**2 + nz1**2)
len2 = math.sqrt(nx2**2 + ny2**2 + nz2**2)

cos_theta = dot / (len1 * len2)
cos_theta = max(-1, min(1, cos_theta))

theta = math.degrees(math.acos(cos_theta))
print(f"{round(theta,2):.2f}")

```

```python
#W8A12
class Complex:
    def __init__(self,real,im):
        self.real=real
        self.im=im
    def multiply(self,other):
        x1=self.real
        x2=self.im
        y1=other.real
        y2=other.im
        return (x1*y1-x2*y2,x1*y2+x2*y1)
r1,i1=map(int,input().split())
a1=Complex(r1,i1)
r2,i2=map(int,input().split())
a2=Complex(r2,i2)
c=a1.multiply(a2)
for k in c:
    print(str(k)+' ',end='')
```

```python
#W8A13
class Triangle:
    def __init__(self,edge1:int,edge2:int,edge3:int):
        self.edge1=edge1
        self.edge2=edge2
        self.edge3=edge3
    def getArea(self):
        a=self.edge1
        b=self.edge2
        c=self.edge3
        if (a+c)<=b or a+b<=c or b+c<=a or a<=0 or b<=0 or c<=0:
            return "invalid"
        p=(a+b+c)/2
        area=(p*(p-a)*(p-b)*(p-c))**0.5
        return f'{round(area,2):.2f}'
a,b,c=map(int,input().split())
tac_giam=Triangle(a,b,c)
print(tac_giam.getArea())

        
```
