```python
#W6A9
a=eval(input())
b=eval(input())
c={}
for aa in a:
    if aa not in c:
        c[aa]=a[aa]
    else:
        c[aa]+=a[aa]
for bb in b:
    if bb not in c:
        c[bb]=b[bb]
    else:
        c[bb]+=b[bb]
d=sorted([k for k in c])
e={}
for k in d:
    e[k]=c[k]
print(e)
```

```python
#W6A12
a=list(map(int,input().split()))
di={}
for k in a:
    if k not in di:
        di[k]=1
    else:
        di[k]+=1
lon='mm'
for key,value in di.items():
    if lon=='mm':
        lon=value
        so=key
        continue
    if lon<value:
        so=key
        lon=value
    elif lon==value and so>key:
        so=key
print(so)
    
```
