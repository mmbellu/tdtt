Hãy nhập thử đầu vào là "a" rồi đọc kỹ Traceback 
để gỡ lỗi nhập liệu. (Dùng try ... except bắt lỗi 
ValueError thêm set)

```python
#W10A3
while True:
    try:
        a=int(input())
        break
    except ValueError:
        print(" a phai la so tu nhien!")
n=str(a)
count=0
di={}
while True:
    tong=0
    for k in n:
        tong+=int(k)**2
    if tong==1:
            print("YES")
            break
    elif tong in di:
         print("NO")
         break
    count+=1
    di[tong]=count
    n=str(tong)
```

Sử dụng try except,kết hợp cùng else... finally.
Tính tổng 2 số, nếu nhập sai kiểu thì thông báo,
luôn luôn in "Kết thúc chương trình." sau khi chạy

```python
#W10A4
try:
    data=list(map(str,input().split()))

    a=float(data[0])
    b=float(data[1])
except IndexError:
    print("ERROR:nhap thieu a hoac b")
except ValueError :
    print("ERROR:nhap sai kieu gia tri cua a hoac b")
else:
    print(f'a+b = {a+b}')
finally:
    print("ket thuc chuong trinh")
```

Viết chương trình chuyển tất cả phần tử của một 
list cho trước về int.
Nếu phần tử nào trong list không thể chuyển được 
thì in cảnh báo và bỏ qua phần tử đó

```python
#W10A5
my_list=list(map(str,input().split()))
for phan_tu in range(len(my_list)):
    try:
        my_list[phan_tu]=int(my_list[phan_tu])
    except ValueError:
        print(f"ERROR:ko the chuyen {my_list[phan_tu]} thanh int")
        
```
