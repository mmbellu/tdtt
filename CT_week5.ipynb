```python
''' W5A1'''
def Max_Of_Two( a, b) -> int:
    return max(a,b)
a,b=map(int,input().split())
print(Max_Of_Two( a,b))
```

```python
'''W5A2'''
def swap(a, b) :
    return b,a
a,b=map(int,input().split())
a,b=swap(a,b)
print(a,b)
```

```python
'''W5A3'''
def isprime(n) -> bool:
    if n==2:
        return True
    for k in range(2,int(n**0.5)+1):
        if n%k==0:
            return False
    return True
print(isprime(int(input())))
```

```python
'''W5A4'''
def so_hoan_hao(n) -> bool:
    tong=1
    for k in range(2,int(n**0.5)+1):
        if n%k==0:
            tong+=k
            tong+=n//k if k**2!=n else 0
    if tong==n:
        return True
    return False
print(so_hoan_hao(int(input())))
```

```python
'''W5A5'''
def tra_index(lis:list,target:int) -> int:
    for k in range(len(lis)):
        if lis[k]==target:
            return k +1
    return -1
lis=list(map(int,input().split()))
target=int(input())
print(tra_index(lis,target))
```

```python
'''W5A6'''
def giai_thua(n):
    tich=1
    for k in range(1,n+1):
        tich*=k
    return tich
print(giai_thua(int(input())))
```

```python
'''W5A7'''
def calc(n:str)->float:
    if '+' in n:
        a,b=map(float,n.split('+'))
        return f'{round(a+b,2):.2f}'
    elif '*' in n:
        a,b=map(float,n.split('*'))
        return f'{round(a*b,2):.2f}'
    elif '-' in n:
        a,b=map(float,n.split('-'))
        return f'{round(a-b,2):.2f}'
    else:
        a,b=map(float,n.split('/'))
        return f'{round(a/b,2):.2f}'
print(calc(input()))
    
```

```python
'''W5A8'''
def kchm(a:int,b:int)->int:
    return bin(a^b).count('1')
a,b=map(int,input().split())
print(kchm(a,b))
```

```python
'''W5A9'''
def tong_cac_chu_so(n:int) -> int:
    tong=0
    while n > 0 :
        tong += n % 10
        n //= 10
    return tong
n=int(input())
print(tong_cac_chu_so(n))
```

```python
'''W5A10'''
#CACH_1:
def dang_cau(a,b):
    dicta={}
    dictb={}
    for k in range(len(a)):
        if a[k] not in dicta:
            dicta[a[k]]=b[k]
        else:
            if b[k]!=dicta[a[k]]:
                return False
    for k in range(len(a)):
        if b[k] not in dictb:
            dictb[b[k]]=a[k]
        else:
            if a[k]!=dictb[b[k]]:
                return False
    for k in dictb:
        if dicta[dictb[k]]!=k:
            return False
    return True
a,b=map(str,input().split())
print(dang_cau(a,b))
```

```python
#CACH2:
'''su dung ham zip(a,b) ham zip se tra ve lan luot cac gia tri a,b tuong ung voi index:
    vi du index 0 se tra ve a0 va b0                                                    '''
def dang_cau(a,b):
    dicta={}
    dictb={}
    for ma,mb in zip(a,b):
        if ma not in dicta:
            dicta[ma]=mb
        else:
            if dicta[ma]!=mb:
                return False
        if mb not in dictb:
            dictb[mb]=ma
        else:
            if dictb[mb]!=ma:
                return False
    return True
a,b=map(str,input().split())
print(dang_cau(a,b))
```
