```python
''' W5A1'''
def Max_Of_Two( a, b) -> int:
    return max(a,b)
a,b=map(int,input().split())
print(Max_Of_Two( a,b))
```

```python
'''W5A2'''
def swap(a, b) :
    return b,a
a,b=map(int,input().split())
a,b=swap(a,b)
print(a,b)
```

```python
'''W5A3'''
def isprime(n) -> bool:
    if n==2:
        return True
    for k in range(2,int(n**0.5)+1):
        if n%k==0:
            return False
    return True
print(isprime(int(input())))
```

```python
'''W5A4'''
def so_hoan_hao(n) -> bool:
    tong=1
    for k in range(2,int(n**0.5)+1):
        if n%k==0:
            tong+=k
            tong+=n//k if k**2!=n else 0
    if tong==n:
        return True
    return False
print(so_hoan_hao(int(input())))
```

```python
'''W5A5'''
def tra_index(lis:list,target:int) -> int:
    for k in range(len(lis)):
        if lis[k]==target:
            return k +1
    return -1
lis=list(map(int,input().split()))
target=int(input())
print(tra_index(lis,target))
```

```python
'''W5A6'''
def giai_thua(n):
    tich=1
    for k in range(1,n+1):
        tich*=k
    return tich
print(giai_thua(int(input())))
```
