```python
#W11A4
def areAnagrams(path1: str, path2: str):
    with open(path1,'r') as file1:
        a=file1.read()
    with open(path2,'r') as file2:
        b=file2.read()
    a=a.replace(' ','')
    b=b.replace(' ','')
    return set(a)==set(b)
path1,path2=map(str,input().split())
print(areAnagrams(path1,path2))
```

f(0)=a
f(1)=b
f(n)=f(n−1)+f(n−2)

```python
#W11A8
n = int(input())
with open("output.txt", 'w') as file:
    f0 = 'a'
    f1 = 'b'
    file.write(f"f(0) = {f0}\n")
    if n >= 1:
        file.write(f"f(1) = {f1}\n")

    for i in range(2, n + 1):
        f0, f1 = f1, f0 + f1
        file.write(f"f({i}) = {f1}\n")
```

```python
#W11A10
def getMoney (path: str) -> list[int]:
    with open(path,'r') as file:
       data = file.read().strip()
       if data == "":
            return []
       return list(map(int, data.split()))
nha=getMoney("test.txt")
def rob(houses: list[int]) -> int:
    if not houses:
        return 0
    if len(houses) == 1:
        return houses[0]

    prev2 = houses[0]
    prev1 = max(houses[0], houses[1])
    for k in range(2,len(houses)):
        curr=max(prev1,prev2+houses[k])

        prev2=prev1
        prev1=curr
    return prev1
print(rob(nha))
```

```python
#W12A13
def moveZeroes (fileName: str) -> list:
    with open(fileName,'r') as file:
        mm=file.readlines()
        count=0
        res=[]
        for num in mm[1::]:
            if num=='0\n':
                count+=1
            else:
                res.append(int(num))
        if count:
            res.extend([0 for i in  range(count)])
        return res
file=input()
print(moveZeroes(file))
```

```python
#W11A14
def averageTime(path: str):
    total_time = 0.0
    count = 0
    with open(path, 'r', encoding='utf-8') as f:
        next(f)  # bỏ dòng tiêu đề
        for line in f:
            parts = line.strip().split(',')
            time = float(parts[1])
            total_time += time
            count += 1

    return round(total_time / count, 2)
```

Để đưa cho khách hàng các lựa chọn, rạp chiếu phim treo các poster về phim cùng với điểm đánh giá IMDB (rating) và mô tả của khách hàng về bộ phim đó.

Hãy viết hàm def findMovies (path: str) -> list[str] giúp khán giả tìm ra phim với điểm đánh giá lớn hơn 8.0 và có mô tả Không phải là "nhàm chán" ("boring").

Hàm nhận đầu vào là đường dẫn đến tệp chứa các poster phim. Hàm trả lại các tên những phim thỏa mãn điều kiện trên. Nếu không tìm thấy phim nào phù hợp, trả về thông báo "All are boring!". Kết quả in theo thứ tự xuất hiện trước sau trong file đầu vào

Tệp có cấu trúc như sau:

id, movie, description, rating
1, Thor, great 3D, 8.0
2, The Rings, Fantasy, 7.5
3, Irish, boring, 6.2
4, Ice song, Fantasy, 8.6
5, Home Alone, Interesting, 6.5

```python
#W11A15
def findMovies (path: str) -> list[str]:
    movie=[]
    with open(path,'r') as file:
        next(file)
        for line in file:
            parts = line.strip().split(',')
            if float(parts[2])>8 and parts[1]!='boring':
                movie.append(parts[0])
    return movie
print(*findMovies(input()))
```

Viết chương trình, nhập vào đường dẫn đến một tệp. Tệp này chứa thông tin của các hình vuông, hình tròn và hình chữ nhật. In ra chu vi của các hình trong tệp trên từng dòng, theo thứ tự xuất hiện trong tệp. Kết quả được làm tròn đến hai chữ số thập phân sau dấu phẩy.
Một tệp ví dụ như sau:

TYPE: SQUARE
SIDE: 5.3

TYPE: RECTANGLE
WIDTH: 5.0 HEIGHT: 16.0

TYPE: CIRCLE
RADIUS: 2.15
Một tệp có thể chứa thông tin của một hoặc nhiều hình và không theo thứ tự nhất định. Mỗi hình sẽ được định nghĩa trên hai dòng, dòng đầu tiên là loại hình (SQUARE, RECTANGLE, CIRCLE) và dòng tiếp theo là các thông số tương ứng với hình đó (SIDE, WIDTH, HEIGHT, RADIUS). Giữa thông tin của hai hình sẽ có một dòng trống.

```python
#W11A16
import math

def main():
    path = input().strip()

    with open(path, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip()]

    i = 0
    while i < len(lines):
        shape_type = lines[i].split(":")[1].strip()
        params = lines[i + 1]

        if shape_type == "SQUARE":
            side = float(params.split(":")[1])
            perimeter = 4 * side

        elif shape_type == "RECTANGLE":
            parts = params.split()
            width = float(parts[1])
            height = float(parts[3])
            perimeter = 2 * (width + height)

        elif shape_type == "CIRCLE":
            radius = float(params.split(":")[1])
            perimeter = 2 * math.pi * radius

        print(f"{perimeter:.2f}")
        i += 2

```

```python
#W11A17
def productExceptSelf(path: str) -> list[int]:
    with open(path, 'r', encoding='utf-8') as f:
        nums = []
        for line in f:
            nums.extend(map(int, line.split()))

    n = len(nums)
    left = [1] * n
    right = [1] * n
    result = [0] * n

    for i in range(1, n):
        left[i] = left[i - 1] * nums[i - 1]

    for i in range(n - 2, -1, -1):
        right[i] = right[i + 1] * nums[i + 1]

    for i in range(n):
        result[i] = left[i] * right[i]

    print(result)
    return result
```

```python
#W11A18
import math

def estimatedTime(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip()]

    # Dòng SRC
    src_parts = lines[0].split()
    x1 = float(src_parts[2])
    y1 = float(src_parts[4])

    # Dòng DES
    des_parts = lines[1].split()
    x2 = float(des_parts[2])
    y2 = float(des_parts[4])

    # Dòng VELOCITY
    v = float(lines[2].split()[1])

    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    time = distance / v

    return round(time, 2)
```

```python
#W11A20
def imageSmoother(path: str):
    with open(path, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip()]

    M = int(lines[0])
    image = []

    for i in range(1, M + 1):
        image.append(list(map(int, lines[i].split())))

    result = [[0] * M for _ in range(M)]

    for i in range(M):
        for j in range(M):
            total = 0
            count = 0

            for dx in (-1, 0, 1):
                for dy in (-1, 0, 1):
                    x = i + dx
                    y = j + dy
                    if 0 <= x < M and 0 <= y < M:
                        total += image[x][y]
                        count += 1

            result[i][j] = total // count

    for row in result:
        print(*row)
```
