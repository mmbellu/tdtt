```python
#W7A3
n=list(map(int,input().split()))
def bubble_sort(n):
    a=len(n)
    for k in range(a):
        swapped = False
        for l in range(a-1):
            if n[l]>n[l+1]:
                n[l+1], n[l]=n[l+1],n[l]
                swapped=True
        if swapped==False:
            break
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
      min_idx = i
      for j in range(i+1, n):
         if arr[j] < arr[min_idx]:
           min_idx = j # Update min_idx
      arr[i], arr[min_idx] = arr[min_idx], arr[i] # Swap
bubble_sort(n)
selection_sort(n)
print(n)
```

```python
#W7A8
'''Bạn được cung cấp: một
danh sách các khoảng thời
gian interval = [[l1, r1], [l2,
r2], ...] và một danh sách
các truy vấn queries = [q1,
q2, ...]
Với mỗi truy vấn q, hãy tìm 
độ dài nhỏ nhất của  
khoảng thời gian [li, ri] sao
cho li ≤ q ≤ ri.                               
Nếu không tồn tại khoảng
thời gian nào như vậy, hãy
trả về -1 cho truy vấn đó.
solve in < O(m.n), hint:
binary search in the
queries list'''
n=eval(input())
a=eval(input())
di={}
b=[]
for k in range(len(n)):
    for o in a:
        if n[k][0]<=o<=n[k][1] :
          if o in di and di[o]>n[k][1]-n[k][0]+1:
              di[o]=n[k][1]-n[k][0]+1
          elif o not in di:
             di[o]=n[k][1]-n[k][0]+1
for i in a:
   b.append(di.get(i,-1))
print(b)
```

```python
#W7A9
'''Cho một danh sách các số
nums, với mỗi phần tử
nums[i], hãy đếm xem có
bao nhiêu phần tử bên phải
nhỏ hơn nums[i].
Trả về danh sách các số
đếm.
Efficient solution: using
heap or binary search tree'''
n=eval(input())
len_n=len(n)
count_greater=[]
for k in range(len_n):
    count=0
    for i in range(k+1,len_n):
        if n[i]<n[k]:
            count+=1
    count_greater.append(count)
print(count_greater)
```

```python
#W7A10
'''Cho một danh sách các số
nguyên nums và một số
nguyên m, hãy tìm tổng
mảng con lớn nhất theo
modulo m.
Tức là, với bất kỳ mảng
con nums[i:j] nào, hãy tính
sum(nums[i:j]) % m và trả                input:[3, 3, 9, 9, 5]
về kết quả lớn nhất có thể.                              7
Gợi ý:
Duy trì tổng tiền tố (prefix               output:6
sum) mod m và sử dụng
tìm kiếm nhị phân để tìm
tổng tiền tố nhỏ nhất chỉ
lớn hơn tổng hiện tại một
chút (để tối đa hóa hiệu số
bao quanh).'''
m=int(input())
n=eval(input())
b=len(n)
maxx=0
for k in range(b):
    for i in range(k+1,b):
        maxx=max(maxx,sum(n[k:i])%m)
print(maxx)
```
