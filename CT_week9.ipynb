```python
#W9A8 
#cach 2
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}
    for m in moves:
        if m == 'R':
            x += 1
        elif m == 'L':
            x -= 1
        elif m == 'U':
            y += 1
        elif m == 'D':
            y -= 1
        if (x, y) in visited:
            return True
        visited.add((x, y))

    return False
print(is_path_crossing(input()))
```
